```python
import spotipy
import spotipy.util as util
import webbrowser
import json
import urllib.request
from nltk.corpus import stopwords
from nltk import pos_tag
import requests
import collections
import nltk
```

```python
with open("news_key.txt", 'r') as key_file:
    news_key = key_file.read()
#Close the file
key_file.close()
print(news_key)
```

```python
country = "us"
url = f"https://newsapi.org/v2/everything?q=country={country}&apiKey={news_key}"
print(url)
```

```python
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
```

```python
news_articles = json.loads(response.read())
```

```python
news_articles
```

```python
#get titles
article_titles = [article['title'] for article in news_articles['articles']]
article_titles
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
tagged_words = pos_tag(article_titles[1].split())
tagged_words
proper_nouns = [ word for word,tag in tagged_words if tag == "NNP"]
print(proper_nouns)
```

```python
keywords = []
for article_title in article_titles:
    tagged_words = pos_tag(article_title.split())
    #collect all the proper nouns
    #use extend to put words together instead of using append (more clean)
    keywords.extend([word for word,tag in tagged_words if tag == "NNP"])

keywords
```

```python
word_counter = collections.Counter(keywords)
word_counter
```
